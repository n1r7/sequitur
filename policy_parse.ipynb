```python
import sys
# !{sys.executable} -m pip install pdfminer
import pdfminer
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
from io import StringIO
import os 

#function that converts PDF to text
#optional parameter PAGES can restrict which pages to process
def convert_pdf_to_txt(path, pages=None):
    if not pages:
        pagenums = set()
    else:
        pagenums = set(pages)
    
    #Instantiate the PDFminer objects
    output = StringIO()
    manager = PDFResourceManager()
    converter = TextConverter(manager, output, laparams=LAParams())
    interpreter = PDFPageInterpreter(manager, converter)
    
    infile = open(path, 'rb') #open the file for read in binary mode
    
    for page in PDFPage.get_pages(infile, pagenums):     #iterate with the pdf interpreter through the pdf pages
        interpreter.process_page(page)
    
    #Close the files and converters
    infile.close()
    converter.close()
    text = output.getvalue()
    output.close()
    
    return text   #return the text as str

text = convert_pdf_to_txt('/Users/n1r7/Desktop/fincrime/pdf_parse/test/Private_equity.pdf') #call the function for the file specified here
# text = split_into_sentences(text)
df = pd.DataFrame({"text":[text]}) #convert the TEXT str to Panda's DF
pd.set_option('display.max_colwidth', 2000)

df
```
