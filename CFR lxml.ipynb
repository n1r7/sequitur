```python
import pandas as pd
import requests
import re
import json
import lxml
from lxml import etree
from io import StringIO, BytesIO
import xmldataset

import pprint
ppsetup = pprint.PrettyPrinter(indent=4)
pp = ppsetup.pprint
```

```python
# Load XML content

url = 'https://www.govinfo.gov/bulkdata/CFR/2019/title-12/CFR-2019-title12-vol1.xml'
page = requests.get(url)
print(page)

doc = page.content
print(type(doc))
```

```python
# Parse content

parser = etree.XMLParser()
tree = etree.parse('/Users/n1r7/Desktop/fincrime/CFR-2019-title-12/CFR-2019-title12-vol1.xml', parser)
root = tree.getroot()
```

```python
contents = []

part_p_elements = tree.xpath('//TITLE//CHAPTER//PART//SECTION//P')
```

```python
for p in part_p_elements:
    content = ' '.join(map(str.strip, p.xpath(".//text()")))
    sectno = ' '.join(map(str.strip, p.xpath("../SECTNO//text()")))
    subjgrp = ' '.join(map(str.strip, p.xpath('../../../HD//text()')))
    subject = ' '.join(map(str.strip, p.xpath("../SUBJECT//text()")))
    cita = ' '.join(map(str.strip, p.xpath("../CITA//text()")))
    part = ' '.join(map(str.strip, p.xpath("../../HD//text()")))
    subpart = ' '.join(map(str.strip, p.xpath('../../../HD//text()')))
    authority = ' '.join(map(str.strip, p.xpath("../../AUTH/P//text()")))
    source = ' '.join(map(str.strip, p.xpath("../../SOURCE/P//text()")))
    lrh = ' '.join(map(str.strip, p.xpath('../../../../LRH//text()')))
    rrh = ' '.join(map(str.strip, p.xpath('../../../../RRH//text()')))
    title = ' '.join(map(str.strip, p.xpath('../../../../CFRTITLE/TITLEHD/HD//text()')))
    contents.append({"Title":title,
                     "LRH":lrh,
                     "RRH":rrh,
                     "Part":part,
                     "Subpart":subpart,
                     "Section Number": sectno, 
                     "Section Subject": subject,
                     "Subject Group":subjgrp,
                     "Statement":content, 
                     "Authority":authority, 
                     "Source":source,
                     "Citation":cita
                    })
```

```python
# pd.set_option('display.max_colwidth',0)
df = pd.DataFrame(data=contents, columns=['LRH','RRH','Title','Source','Authority','Part', 'Subpart',
                                          'Section Number','Section Subject','Subject Group', 
                                          'Statement','Citation'])
df.tail(2000)
```
